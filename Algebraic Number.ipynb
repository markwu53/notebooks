Definition. Let $F\subseteq K$ be two fields. $K$ is called a field extension of $F$, denoted $K/F$.

Remark.
Consider $K$ as an $F$-algebra. For each $\alpha\in K$, define a $F$-algebra homomorphism $\phi_\alpha:F[x]\to K$ by sending $x\mapsto \alpha$ and $1\mapsto 1$. Since $F[x]$ is a principal ideal domain, if the kernel of $\phi_\alpha$ is not zero, then denote a generator of the ideal by $p_\alpha$. We can also write $p_\alpha(\alpha)=0$. In this case we say $\alpha$ is algebraic over $F$, and $p_\alpha$ a minimal polynomial of $\alpha$, and the degree of $p_\alpha$ the degree of $\alpha$.

Definition.
* Let $F\subseteq K$ be two fields. $K$ is called a $\textbf{field extension}$ of $F$, denoted $K/F$.
* Let $K/F$ be a field extension. If $K$ is of finite dimension as a $F$-vector space, then $K$ is called a $\textbf{finite extension}$ of $F$. The dimension of $K$ as a $F$-vector space is called the $\textbf{degree}$ of the extension.
* Let $K/F$ be a field extension. An element $\alpha\in K$ is called $\textbf{algebraic}$ over $F$ if there exists a polynomial $p\in F[x]$ such that $p(\alpha)=0$.
* Let $K/F$ be a field extension. Let $\alpha\in K$ be algebraic over $F$. The minimal degree of polynomial $p$ such that $p(\alpha)=0$ is called the $\textbf{degree}$ of $\alpha$. The polynomial $p$ that has degree equal to the degree of $\alpha$ and leading coefficient $1$ is called the $\textbf{minimal polynomial}$ of $\alpha$.
* Let $K/F$ be a field extension. If every element in $K$ is algebraic over $F$, then $K$ is called an $\textbf{algebraic extension}$ of $F$.
* Let $K/F$ be a field extension. Let $G\subseteq K$ be a finite subset. Denote $F[G]$ the $F$-algebra generated by $G$. Denote $F(G)$ the field generated by $F$ and $G$, which is also the field of fraction of $F[G]$. If $G=\{ \alpha_1,\alpha_2,\cdots,\alpha_n\}$, we also write $F[G]=F[\alpha_1,\alpha_2,\cdots,\alpha_n]$ and $F(G)=F(\alpha_1,\alpha_2,\cdots,\alpha_n)$.
* Let $F$ be a field. If $F$ has no proper algebraic extension, then $F$ is $\textbf{algebraically closed}$. The smallest algebraically closed field that contains $F$ is called the $\textbf{algebraic closure}$ of $F$.
* An $\textbf{algebraic number field}$ is a finite field extension over $\mathbb{Q}$.

Remark.
* Every field has an algebraic closure by Zorn's lemma.
* The algebraic closure of a field is unique.
* Every algebraic extension of a field can be embedded into the algebraic closure of the field.
* If $F$ is an alegraically closed field, then the only irreducible polynomial in $F[x]$ are linear polynomials. Otherwise, suppose $p$ is irreducible with degree more than $1$, consider $F[x]/(p)$, which is a field contains $F$, and $x\in F[x]/(p)$, $x\notin F$ and is algebraic over $F$, contradiction.
* The minimal polynomial of an algebraic elemnt is unique.
* A finite extension is an algebraic extension.
* Let $K/F$ be a field extension. If $\alpha\in K$ is algebraic over $F$, then $F(\alpha)=F[\alpha]$, and is a finite extension of $F$ and therefore, an algebraic extension of $F$.
* Let $L/K/F$ be field extentions. If $L$ is algebraic over $F$ then $L$ is algebraic over $K$.
* If $K/F$ and $L/K$ are finite field extensions, then $L/K$ is a finite extension.
* If $K/F$ and $L/K$ are algebraic field extensions, then $L/K$ is an algebraic extension.
* $F[\alpha_1,\alpha_2]=F[\alpha_1][\alpha_2]=F[\alpha_2][\alpha_1]$. $F(\alpha_1,\alpha_2)=F(\alpha_1)(\alpha_2)=F(\alpha_2)(\alpha_1)$.
* Let $K/F$ be a field extension. The set of elements that are algebraic over $F$ is a field.
* $\mathbb{Q}(\sqrt{2})=\mathbb{Q}[\sqrt{2}]$.

Remark.
Let $K/F$ be an algebraic field extension. Given $\alpha\in K$, $\alpha$ annihialates an ideal of $F[x]$ which is a principal ideal generated by the minimal polynomial of $\alpha$.

Definition.
* Let $A\subseteq B$ be an inclusion of commutative rings. $B$ is called a ring extension of $A$, denoted $B/A$.
* Let $A$ be a commutataive ring. A $\textbf{monic polynomial}$ is a polynomial in $A[x]$ with leading coefficient $1$.
* Let $B/A$ be an ring extension. An element $\alpha\in B$ is $\textbf{integral over}$ $A$ if there is a monic polynomial $p\in A[x]$ such that $p(\alpha)=0$.
* Let $B/A$ be a ring extension. $B$ is an $\textbf{integral extension}$ over $A$ if every element in $B$ is integral over $A$.

Remark.
We want to define a concept of *integrally closed* ring and *integral closure* similar to algebraically closed field and algebraic closure. But Zorn's lemma doesn't work in this setting. We have to consider the concepts when there is a containing ring.

Definition.
* Let $B/A$ be a ring extension. $A$ is $\textbf{integrally closed in}$ $B$ if $A$ has no proper integral extension in $B$.
* Let $B/A$ be a ring extension. The $\textbf{integral closure}$ of $A$ in $B$ is the smallest integral extension of $A$ in $B$ that is integrally closed.
* Let $A$ be an integral domains. $A$ is $\textbf{integrally closed}$ if $A$ is integrally closed in its field of fraction.
* Let $B/A$ be a ring extension of unique factorization domains. $B/A$ is a $\textbf{finite extension}$ if $B$ is of finite rank as an $A$-module.
* Let $B/A$ be a ring extension. Denote $\mathcal{O}(B,A)$ the set of elements in $B$ that are integral over $A$.
* Let $F$ be an algebraic number field. Denote $\mathcal{O}_F$ to be $\mathcal{O}(F,\mathbb{Z})$.

Remark.
* $\mathcal{O}(B,A)$ is a ring. Therefore, it's the integral closure of $A$ in $B$.
* A finite extension is an integral extension.
* If $B/A$ and $C/B$ are finite extension, then $C/A$ is finite extension.
* If $B/A$ and $C/B$ are integral extension, then $C/A$ is integral extension.
* Let $B/A$ be a ring extention. Let $\alpha\in B$ be an integral element. Then $A[\alpha]$ is an integral extension of $A$.
* Let $B/A$ be a ring extention. Let $\alpha,\beta\in B$. Then $A[\alpha,\beta]=A[\alpha][\beta]=A[\beta][\alpha]$.
* $\mathcal{O}_F$ is a ring. 

Definition.
* An integral domain is $\textbf{absolutely integrally closed}$ if it has no proper integral extension.
* The $\textbf{absolute integral closure}$ of an integral domain $A$ is an absolutely integrally closed domain which is integral over $A$.

Remark.
* An algebraically closed field is an absolutely integrally closed domain.
* The integral closure of a domain in the algebraic closure of its field of fraction is its absolute integral closure.

Definition.
$\mathcal{O}_F$ is called the $\textbf{ring of integer}$ of $F$.

Remark.
* $\mathbb{Z}\subseteq \mathcal{O}_F$.
* $F$ is the field of fraction of $\mathcal{O}_F$.
