Let $R$ be an integral domain. Let $F$ be the field of fraction of $R$. Let $K$ be a field extension of $F$. Let $R[x]$ and $F[x]$ be the polynomial ring with coefficients in $R$ and $F$ respectively.

Definition. A *monic* polynomial is an element in $R[x]$ with leading coefficient $1$. A *primitive* polynomial is an element in $R[x]$ such that the coeffecients of the polynomial have no non-unit common factors.

Remark. A monic polynomial is primitive. A primitive polynomial divided by the leading coefficient becomes a *monic* polynomial in $F[x]$. Any polynomial in $R[x]$ can factor out common factors in the coefficients and becomes a primitive polynomial. Any polynomial in $F[x]$ can clear the denominator (by multiplying common multiple of the coefficients) and become a polynomial in $R[x]$.

Definition. An element $\alpha$ in $K$ is said to be *algebraic* over $F$ or $R$ if there is a polynomial $p(x)$ in $F[x]$ such that $p(\alpha)=0$. An element $\alpha$ in $K$ is said to be *integral* over $F$ or $R$ if there is a *monic* polynomial $p(x)$ in $R[x]$ such that $p(\alpha)=0$.

Remark. The polynomial in above definition is not unique. But if we consider monic polynomial with least degree, then it is unique.

Definition. The polynomial in above remark is called the *minimal* polynomial.

Remark. A minimal polynomial is necessarily an irreducible.

Remark. An integral element is an element that satisifies an integral polynomial (the coefficients of the minimal polynomial are integral). An algebraic element is an element that satisfies a fractional polynomial (the coefficients of the minimal polynomial are fractional).

Remark. Elements in $R$ are always integral over $R$ or $F$. Elements in $F$ are always algebraic over $R$ or $F$.

Definition. $R$ is said to be *integrally closed* if elements in $F\backslash R$ are not integral over $R$.

Definition. $F$ is said to be *algebraically closed* if every polynomial in $F[x]$ has a root in $F$.

Claim. The set of elements integral over $R$ forms a ring. The set of elements algebraic over $F$ forms a field.

Proof. Let $\alpha$ and $\beta$ be two elements that are integral over $R$ with minimal polynomial $P(x)$ and $Q(x)$. Then for any $r\in R$, $r^nP(x/r)$ is the minimal polynomial of $r\alpha$ where $n$ is the degree of $P(x)$. Let $M(z)$ be the resultant of $P(x)$ and $Q(z-x)$. Then $M(x)$ is the minimal polynomial of $\alpha+\beta$. Let $N(z)$ be the resultant of $P(x)$ and $x^nQ(z/x)$. Then $N(x)$ is the minimal polynomial of $\alpha\beta$. The minimal polynomial of $1/\alpha$ is $x^nP(1/x)$ divided by the leading coefficient of the resulting polynomial.///

Claim. $F[\alpha]=F(\alpha)$. $F[\alpha]$, as a $F$-vector space, has dimension equal to the degree of $m(x)$. $F[\alpha]$ is a finite extension of $F$. Proof. Consider the map $\phi:F[x]\to F[\alpha]$. The kernel is the ideal generated by the minimal polynomial $m(x)$. It's prime and maximal. Therefore, $F[\alpha]$ is a field.///

Remark. $F[\alpha,\beta]=F[\alpha][\beta]=F[\beta][\alpha]$. If $F[\alpha], F[\beta]$ over $F$ have degree $d_\alpha$ and $d_\beta$, then $F[\alpha,\beta]$ over $F$ has degree at most $d_\alpha d_\beta$.

Claim. If $K$ is algebraic over $F$ and $L$ is algebraic over $K$, then $L$ is algebraic over $F$. Proof. Suppose $\alpha$ is an element in $L$ with the minimal polynomial $m(x)$ of degree $d$ in $K[x]$. Then $m(x)$ has at most $d+1$ coefficients in $K$. Suppose they are $\{\beta_0,\beta_1,\cdots,\beta_d\}$. Then $\alpha$ is algebraic over $F[\beta_0,\beta_1,\cdots,\beta_d]$. So $\alpha$ is algebraic over $F$.///

Definition. Let $F\subseteq K$ be a field extension. If $K$ as a $F$-vector space has finite dimension, then we say $K$ is a *finite* extension of $F$, and the dimension is called the *degree$ of the extension.

Remark. A finite degree field extension is an algebraic extension.

Let $R$ be a unique factorization domain. Let $\alpha$ be an element in $K$ that is algebraic over $F$ with minimal polynomial $m(x)$. Let $D_\alpha$ be the set of irreducibles that appear in the denominators of the coefficients of $m(x)$. It is also equal to the set of irreducibles that appear in the factorization of the leading coefficient of the primitive polynomial corresponding to $m(x)$. Let $S$ be a set of elements that are algebraic over $F$. Let $D_S$ be the union of all $D_\alpha$ for $\alpha\in S$.

Claim. Let $G=\{\alpha_1,\alpha_2,\cdots,\alpha_n\}$ be a finite subset of $K$ that are algebraic over $F$. Let $R[G]$ be the $R$-algebra generated by $G$. Then $D_{R[G]}=D_G$.

Proof. By the construction of the minimal polynomial of addition and multiplication of two algebraic elements (via resultant), we see that the process does not change the set of irreducibles that appear in the denominators of the coefficients of the involved minimal polynomials.///

Claim. Let $k$ be a field. Let $K$ be a field extension of $k$. Let $G=\{\alpha_1,\alpha_2,\cdots,\alpha_n\}$ be a finite subset of $K$. If $K=k[G]$, then $K$ is algebraic over $k$.

Proof. For the case $n=1$, define a map $\phi:k[x]\to k[\alpha_1]$ by $x\mapsto\alpha_1$. The image of the map is a field, so the kernel of the map is a maximal ideal, and is generated by an irreducible monic polynomial $m(x)$ since $k[x]$ is a principal ideal domain. $m(x)$ is the minimal polynomial of $\alpha_1$.

Let's prove the claim by induction on the cardinality of $G$. Suppose the claim is true for all $|G|<n$. We want to prove that the claim is true for $|G|=n$. Let $R=k[\alpha_1]$ and $F=k(\alpha_1)$. Let $G'=\{\alpha_2,\cdots,\alpha_n\}$. Then $K=R[G']$. By induction assumption, $K$ is algebraic over $F$. By the previous claim, we have $D_K=D_{G'}$, in particular, it's a finite set. But $F\subseteq K$, so $D_F$ is also finite. If $\alpha_1$ is not algebraic over $k$, then $k[\alpha_1]$ is isomorphic to $k[x]$. Then $D_F$ cannot be finite. So $\alpha_1$ must be algebraic over $k$. Using this method, we see that every $\alpha_i$ is algebraic over $k$. So $K$ is algebraic over $k$.///

Remark. In the above claim, $K$ is also a finite degree extension of $k$.

Remark. In the above claim, if we assume $k$ is an algebraically closed field, then $K=k$.

Claim. Let $F$ be a field. Then $F[x]$ has infinitely many irreducibles.

Proof. Suppose $F[x]$ only have finitely many irreducibles. Let $\{p_1,p_2,\cdots,p_n\}$ be the set of irreducibles. Consider $p=p_1p_2\cdots p_n+1$. It cannot be factored into irreducibles. Therefore, it is an irreducible. Contradiction.///
