Statement. Let $A\subseteq B$ be an ring extension of integral domains. Let $\alpha,\beta\in B$. If $\alpha$ and $\beta$ are integral over $A$, then $\alpha\beta$ and $\alpha+\beta$ are integral over $A$.

Proof.

Let 
$$a=x^n+a_{n-1}x^{n-1}+\cdots+a_0$$
$$b=x^m+b_{m-1}x^{m-1}+\cdots+b_0$$
be the minimal monic polynomial of $\alpha$ and $\beta$.

Consider the $A$-submodule $M$ of $A[x,y]$ generated by $\mathcal{B}=\{1,\cdots,x^iy^j,\cdots\}$ where $0\le i< n$ and $0\le j< m$. This is a free $A$-submodule of rank $mn$. Each polynomial in $M$ has degree at most $mn-1$.

Let 
$$\phi_x:A[x]\to A[x,y], x\mapsto x$$
$$\phi_y:A[x]\to A[x,y], x\mapsto y$$
Let $p=\phi_x(a)\phi_y(b)\in A[x,y]$. Consider the quotient $R=A[x,y]/I$ where $I$ is the principal ideal generated by $p$. For each element in $R$ we can choose a representation element in $M$. This defines a function, not a homomorphism, $\psi_p:A[x,y]\to V$.

Let $p=x^ny^m+r$, so $r\in V$, and $r\neq0$. Let $\omega=xy\in A[x,y]$. We want to prove that the following set also generates $V$.
$$K=\{1,\psi_p(\omega),\cdots,\psi_p(\omega^i),\cdots,\psi_p(\omega^{mn-1})\}$$.

Another way, choose the first $k$ such that
$$K=\{1,\psi_p(\omega),\psi_p(\omega^2),\cdots,\psi_p(\omega^k)\}$$
becomes linearly dependent. This means that the nontranscendental elements forms a ring. We want to prove that the leading coefficient can be choosen to be $1$, so it's a monic polynomial.

We have shown that the set $N$ of elements nontranscendental over $A$ and the set $M$ of elements integral over $A$ are both subring of $B$, and $A\subseteq M\subseteq N\subseteq B$.
