Remark. Given a set, if there is a property that we can describe for an element, and we can test if an element has such property, then this property is a *testable property*. We can use a testable property to select a unique subset of a set.

Remark. Given a set $X$ and a subset $G$, we can talk about the subset of $X$ generated by $G$ with a certain property $P$. Define property $P_2$ to be the property on a subset of $X$ such that it contains $G$ and it has the property $P$. The property $P_2$ defines a subset of the power set of $X$. Then we take the intersection of it. It is defined and unique.

Review of topology. Given a set $X$. An *open topology* $\mathcal{O}$ is a subset of the power set of $X$, such that
* $X$ and the empty subset $\phi$ are  in $\mathcal{O}$.
* $\mathcal{O}$ is closed under (arbitrary) union and finite intersection.

A *topology* $\mathcal{T}$ on $X$ is a subset of the power set of $X$, generated by an open topology, and is closed under complement, finite union and intersection. The sets in the open topology are *open* sets. The complements of open sets are *closed* sets. A *base* of a topology is a subset of the open topology such that every open set can be written as a union of sets in the base.

Remark. An open topology is a set of subsets of $X$, which is a subset of the power set of $X$, or an element in the power set of the power set of $X$. The definition of open topology used the method of a testable property. The two conditions above define a testable property that we can use to select the set of open topologies, which is a subset of the power set of the power set of $X$. Then an open topology is one element in the set. A topology is also a subset of the power set of $X$, and is defined by subset generation method. A base is one element in the set of basis, which is defined using the property method as a subset of the power set of the open topology of a topology.

Notation. Let $R$ be a communtative ring. Let $\mathcal{I}$ be the set of ideals of $R$. Let $\mathcal{I}^-$ be the set of proper ideals of $R$.

Remark. Consider an ideal as a $R$-submodule. Define addition on $\mathcal{I}$ by letting $I_1+I_2$ to be the direct sum of $R$-submodules. Define multiplication on $\mathcal{I}$ by letting $I_1I_2$ to be the tensor product of $R$-submodules. This turn $\mathcal{I}$ into a semi-ring with zero ideal as the additional zero and $R$ itself as the multiplication identity.

Remark. $I_1+I_2$ is equal to the ideal generated by $I_1$ and $I_2$, which is by definition the smallest ideal containing $I_1$ and $I_2$. For a subset $\mathcal{J}$ of $\mathcal{I}$, $\cap\mathcal{J}$ is the ideal contained in all of $\mathcal{J}$, and $\oplus\mathcal{J}$, direct sum of $\mathcal{J}$, is the ideal generated by $\mathcal{J}$.

Definition. The *spectrum* of $R$, denoted $Spec(R)$, is the set of prime ideals of $R$.

Remark. The spectrum of $R$ is a subset of $\mathcal{I}$. It doesn't have a algebraic structure but it can be given a topological structure. In fact, $\mathcal{I}$ can be given a topological structure, and $Spec(R)$ has the subspace topology.

Definition. Define a function $T_C:\mathcal{I}\to P(\mathcal{I})$, where $P(\mathcal{I})$ is the power set of $\mathcal{I}$, such that for an ideal $I$, $T_C(I)=\{J\in\mathcal{I}: I\subseteq J\}$. Define a function $T_O$ such that $T_O(I)$ is the complement of $T_C(I)$ in $\mathcal{I}$. 

Remark. For a subset $\mathcal{J}$ in $\mathcal{I}$, $\cap_{I\in\mathcal{J}} T_C(I)=T_C(\oplus \mathcal{J})$, and if $\mathcal{J}$ is finite, $\cup_{I\in\mathcal{J}} T_C(I)=T_C(\cap \mathcal{J})$. This shows that $T_C(\mathcal{I})$, the image of $T_C$, is closed under intersection and finite union. Also, $T_O(\mathcal{I})$, the image of $T_O$, is closed under union and finite intersection.

Definition. Using $T_C(\mathcal{I})$ as the set of closed sets, or equivalently, using $T_O(\mathcal{I})$ as the set of open sets, throwing in empty set, we can define a topology on $\mathcal{I}$. This is the *Zaraski topology* on $\mathcal{I}$. Restricting on $Sepc(R)$ becomes the *Zaraski topology* on $Spec(R)$.

Remark. For an ideal $I$, $T_C(I)=\cap T_C(P_f)$ and $T_O(I)=\cup T_O(P_f)$ where $P_f$ is the principal ideal generated by $f\in I$.

Remark. $T_C$ and $T_O$ are not injective if restrict the codomain to $Sepc(R)$.

Definition. Let $R$ be a commutative ring. A *multiplicative set* $S$ is a subset of $R$ that is closed under multiplication and contains $1$.

Remark. A multiplicative set is a submonoid of a ring considered as a multiplicative monoid.

Remark. The set of multiplicative sets of a ring forms a space.

Definition. Given $f\in R$, the set $\{1,f,f^2,\cdots\}$ is a multiplicative set, denoted $M_f$.

Definition. Let $R$ be an integral domain. Define a function $L$ from $R$ to the set of subrings of the field of fraction of $R$, called *localization*, such that $L(f)=R/M_f$.

Definition. $L$ can be extended to the domain of the set of ideals of $R$, by $L(I)=\cap L(f)$ where $f$ runs in $I$.

Remark. The functions $L$ and $T_O$ induce a function $\mathcal{O}$ from the set of open sets of $Spec(R)$ to the set of subrings of the field of fraction of $R$ by $\mathcal{O}(A)=L(I)$ where $I$ is an ideal such that $T_O(I)=A$ for an open set $A$ in $Spec(R)$. Proof. We need to show that if $T_C(I_1)=T_C(I_2)$ then $L(I_1)=L(I_2)$.

Definition. Given $f\in R$, let $D_f$ be the subset of $Spec(R)$ such that each element does *not* contain the principal ideal $(f)$.

Remark. Such $D_f$ form a base of the Zaraski topology.

Remark. The complement of any proper prime ideal is a multiplicative set.

Definition. Given $f\in R$, let $S$ be the multiplicative set generated by $f$. The field of fraction $R/S$ is called the *localization* of $R$ at $f$, denoted $R_f$.

Definition. Let $R$ be a commutative ring. Let $X=Spec(R)$. Let $\mathcal{O}_X$ be the presheaf from the poset of open sets of $Spec(R)$ to the category $Ring$, such that for $f\in R$, $D_f$ is mapped to $R_f$, the localization of $R$ at $f$.
