Let $(X,\mathcal{T}_X)$ be a topological space.

Definition.
* Let $[n]$ denote the set $\{0,1,\cdots,n\}$ for nonnegative integer $n$.
* The $\textbf{standard n}$-$\textbf{simplex}$, denoted $\Delta^n$, is the set of functions $p:[n]\to\mathbb{R}$ such that $p\ge 0$ and $\sum_{x\in[n]}p(x)=1$. $\Delta^n$ is identified with the subset of the Euclidean space $\mathbb{R}^{n+1}$ given by $\{x\in\mathbb{R}^{n+1}:x_0+x_1+\cdots+x_n=1, x_i\ge 0\}$ with the subspace topology.
* Let $F_n$ be the set of injective and order preserving functions from $[n-1]$ to $[n]$.
* Define a function $s:F_n\to[n]$ such that $s(f)$ is the number $i\in[n]$ that is not in the image of $f$.
* For each $f\in F_n$ define a map, called $\textbf{face map}$, $\phi_f:\Delta^{n-1}\to\Delta^n$ by $\phi_f(q)=p$ such that $p(f(i))=q(i)$ and $p(s(f))=0$.
* By composing fact maps, we can define maps $\Delta^{m}\to\Delta^{n}$ for integers $0\le m<n$. Among them we have $\textbf{front face map}$ $\alpha_{m,n}$ that maps $[m]$ to first $(m+1)$ indices in $[n]$, and $\textbf{back face map}$ $\beta_{m,n}$ that maps $[m]$ to last $(m+1)$ indices in $n$.

Remark.
* $F_n$ has $n+1$ elements.
* A face map is an injective map. The image of a face map is once face/edge of the standard simplex.

Definition.
* A $\textbf{singular n}$-$\textbf{simplex}$ in $X$ is a continuous map $\sigma:\Delta^n\to X$.
* Let $C_n(X)$ be the free abelian group generated by the set of singular n-simplex in $X$. Each element of $C_n(X)$ is called a $\textbf{singular n}$-$\textbf{chain}$.
* For a singular $n$-simplex $\sigma$ in $X$ define an element in $C_{n-1}(X)$ to be $\sum_{f\in F_n}(-1)^{s(f)} \sigma\circ\phi_f$. This defines a group homomorphism $\partial_n:C_n(X)\to C_{n-1}(X)$, called the $\textbf{boundary map}$ for $n\ge 1$. Define $\partial_0$ to be $\partial_0:C_0(X)\to 0$.
* $(C_n(X),\partial_n)_n$ or $(C_n(X))_n$ is called the $\textbf{singular complex}$ or the $\textbf{chain complex}$ of abelian groups.
* The kernel of $\partial_n$ are called $n$-$\textbf{cycles}$, the image of $\partial_{n+1}$ are called $n$-$\textbf{boundaries}$.

Remark.
* An element in a free abelian group is a finite sum of the generators.
* An element in $C_n(X)$ is a combination of several $n$-dimensional hypersurfaces. We want to know whether they can enclose a $(n+1)$-dimensional block (in $1$-dim it's a cycle), and whether this block is whole block without holes. The boundary map is designed so that if they enclose a block, each hypersurface is used two times by two adjacent hypersurfaces, and the sign is designed so that they cancel each other. So the boundary of a cycle is zero. A whole block's boundary is always a cycle, but a cycle may not enclose a whole block.
* $\partial_n\circ\partial_{n+1}=0$.

Definition.
* The $n$-th $\textbf{homology group}$ of $X$, denoted $H_n(X)$, is the quotient group $Ker(\partial_n)/Im(\partial_{n+1})$. An element of $H_n(X)$ is called a $\textbf{homology class}$.
* In the construction of $C_n(X)$, we can think of $C_n(X)$ being a free $R$-module with $R=\mathbb{Z}$. If we allow $R$ to be a general commutative ring, we get $\textbf{homology group with coefficient}$ in $R$, $H_n(X;R)$.
* If $Y\subseteq X$ is a subspace with subspace topology, $C_n(X,Y)=C_n(X)/C_n(Y)$ and the boundary operator $\partial_n$ form a chain complex $(C_n(X,Y),\partial_n)_n$. The homology of this chain complex is called $\textbf{relative homology}$, denoted $H_n(X,Y)$.

Remark.
* For a path connected topological space $X$, $Im(\partial_1)$ is generated by finitely many points such that the sum of the coefficients is zero. $Ker(\partial_0)=C_0(X)$ can be partitioned according to the sum of the coefficients of an element. So $H_0(X)=Ker(\partial_0)/Im(\partial_1)=\mathbb{Z}$.
* A contractible space has all homology group $H_n(X)=0$ for $n\ge 1$.
* Homology is a homotopy invariant, or topological invariant.

Definition. 
* Let $A$ be an abelian group. Let $C_n^{*}(X)=Hom(C_n(X),A)$. The dual of $\partial_n$ is $d_{n-1}:C_{n-1}^{*}(X)\to C_n^{*}(X)$. An element in $C_n^{*}(X)$ is called a $\textbf{singular}$ $n$-$\textbf{cochain}$. The collection $(C_n^{*}(X),d_n)_n$ is called a $\textbf{cochain complex}$. Elements in the kernel of $d_n$ are called $\textbf{cocycles}$. Elements in the image of $d_n$ are called $\textbf{coboundaries}$. 

* $H^n(X,A)=Ker(d_n)/Im(d_{n-1})$ is called the $n$-th $\textbf{cohomology group}$. An element in $H^n(X,A)$ is called a $\textbf{cohomology class}$. We write $H^n(X,\mathbb{Z})$ as $H^n(X)$.
* We can replace $A$ with a commutative ring $R$ or an $R$-module $M$, then we get the cohomology $H^n(X,R)$ or $H^n(X,M)$ which are $R$-modules.

Remark.
* $d_n\circ d_{n-1}=0$.
* If $A$ is a commutative ring, $Hom(C_n,A)$ is an $A$-module. We tempt to say it's a ring or an $A$-algebra, but it's not. Suppose we define $f.g(x)=f(x)g(x)$, then since $f,g$ and $f.g$ are group homomorphisms $f.g(2x)=2f.g(x)=2f(x)g(x)$ but on the other hand $f.g(2x)=f(2x)g(2x)=4f(x)g(x)$.

Definition. Let $(Y,\mathcal{T}_Y)$ be another topological space. Let $f:X\to Y$ be a continuous function. Given a singular $n$-simplex $\sigma$ in $X$, and a $n$-cochain $\phi$ in $C_n^{*}(Y)$,
* The mapping $\sigma\mapsto f\circ\sigma$ defines a $\textbf{pushforward}$ function $f_{*}:H_n(X)\to H_n(Y)$.
* The mapping $\phi\mapsto\theta$ where $\theta(\sigma)=\phi(f\circ\sigma)$ defines a $\textbf{pullback}$ function $f^{*}:H^n(Y,A)\to H^n(X,A)$.

Remark.
* To show pushforward and pullback are well defined, we observe that $f_{n-1}\circ\partial_n=\partial_{n-1}\circ f_n$ where $f_n:C_n(X)\to C_n(Y)$ is the group homomorphism induced by $f$, and $f^{n+1}\circ d_{n}=d_{n}\circ f^{n}$ where $f^n:C^n(Y)\to C^n(X)$ is the group/module homomorphism induced by $f$.

Definition. Let $p,q\ge 0$ and $n=p+q$ be integers. Let $R$ be a commutative ring.
* Given a singular $n$-simplex $\sigma$ and a $p$-cochain $\phi$ in $X$, $(\sigma,\phi)\mapsto\phi(\sigma\circ\alpha_{p,n})\cdot(\sigma\circ\beta_{q,n})$ defines a bilinear map, called $\textbf{cap}$, $\cap:H_n(X;R)\times H^p(X;R)\to H_q(X;R)$.
* Given a singular $n$-simplex $\sigma$, a $p$-cochain $\phi$ and a $q$-cochain $\psi$ in $X$, $\phi\cup\psi (\sigma)=\phi(\sigma\circ\alpha_{p,n})\cdot \psi(\sigma\circ\beta_{q,n})$ defines a bilinear map, called $\textbf{cup}$, $\cup:H^p(X,R)\times H^q(X,R)\to H^n(X,R)$.
