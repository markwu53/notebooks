Defintion.
A $\textbf{symplectic structure}$ on a $\mathbb{R}$-vector space $W$ is a *nondegenerate* *alternating* bilinear function $Q:W\times W\to \mathbb{R}$.

Remark.
* Alternating bilinear function means $Q(x,y)=-Q(y,x)$.
* Nondegenerate means for any $x\in W$ there exists a $y\in W$ such that $Q(x,y)\ne0$.
* Nondegenerate also menas $Q$ is an isomorphism if viewed as $Q:W\to (W\to\mathbb{R})$ or $Q:W\to W^*$.

Definition. Let $V$ be a finited dimensional $\mathbb{R}$-vector space.
* Let $V^n=V\otimes\cdots\otimes V$ be the tensor product. Let $T(V)=\oplus_{n=0}^\infty V^n$ be the $\textbf{tensor algebra}$ of $V$.
* Let $I$ be the ideal of $T(V)$ generated by elements of the form $v\otimes v$ for $v\in V$. Let $\bigwedge V$ be the quotient algebra $T(V)/I$, called the $\textbf{exterior algebra}$ of $V$. The multiplication in the exterior algebra is denoted by $\wedge$.

Remark.
* We have $v\wedge v=0$ true for $v\in V$, but not necessarilly true for a general element $\alpha\in\bigwedge V$. For example in a vector space with basis $\{e_1,e_2,e_3,e_4\}$, a symplectic structure that corresponding to a two form $\omega=e_1^*\wedge e_3^*+e_2^*\wedge e_4^*$ does not have $\omega\wedge\omega=0$. In fact, $\omega\wedge\omega$ is a number times the volume form.
