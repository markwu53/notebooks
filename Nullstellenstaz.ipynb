Proposition. Let $k\subseteq K$ be a field extension. Let $\alpha$ be an element in $K$. If $K=k[\alpha]$, then $K$ is algebraic over $k$.

Proof. Define $\phi:k[x]\to K$ by $x\mapsto \alpha$. Let $\mathfrak{a}$ be the kernel. $\mathfrak{a}$ is a maximal ideal, and in particular, a proper ideal. Since $k[x]$ is a principal ideal domain, $\mathfrak{a}=(p)$. We have $p(\alpha)=0$. So $\alpha$ is algebraic over $k$. Therefore, $K$ is algebraic over $k$.///

Claim. Let $R$ be a unique factorization domain. Let $F$ be the field of fraction of $R$. Let $K$ be a field extention of $F$. We have $R\subseteq F\subseteq K$. Let $\alpha$ be an element in $K$ that is algebraic over $F$. Let $D_\alpha=\{p_1,p_2,\cdots,p_m\}$ be the set of irreducibles that appear in the denominators of the coefficients of the minimal polynomial of $\alpha$. Then
$$
D_{R[\alpha]}:=\cup_{\beta\in R[\alpha]} D_\beta = D_\alpha
$$
where $R[\alpha]$ is the $R$-algebra generated by $\alpha$. Also,
$$
D_{R[\alpha_1,\alpha_2,\cdots,\alpha_n]}
:=\cup_{\beta\in R[\alpha_1,\alpha_2,\cdots,\alpha_n]} D_\beta
=D_{\alpha_1}
\cup D_{\alpha_2}
\cup \cdots
\cup D_{\alpha_n}
$$

Proposition. Let $k\subseteq K$ be a field extension. Let $\alpha_1,\alpha_2,\cdots,\alpha_n$ be elements in $K$. If $K=k[\alpha_1,\alpha_2,\cdots,\alpha_n]$, then $K$ is algebraic over $k$.

Proof. We have proved this for $n=1$. Suppose the statement is true for the case when the number of generators is less than $n$. Let's prove the statment for the case when the number of generators is equal to $n$. Let $k'=k(\alpha_1)$, the field generated by $\alpha_1$ over $k$. It is contained in $K$. It is also the field of fraction of $k[\alpha_1]$. We see that $K=k'[\alpha_2,\cdots,\alpha_n]$. So by induction assumption, $K$ is algebraic over $k'$. By previous claim, 
$D_K=
D_{\alpha_2}
\cup \cdots
\cup D_{\alpha_n}$
is finite. This includes all elements in $k'=k(\alpha_1)$. So $D_{k'}$ is finite. On the other hand, if $\alpha_1$ is not algebraic over $k$. Then it's transcendental. Then $k(\alpha_1)$ is isomorphic to $k(x)$, so $D_{k'}$, the set of irreducible denominators over $k[\alpha_1]$ is infinite, because every irreducible can be in the denominator, and there are infinitely many of them. So $\alpha_1$ must be algebraic over $k$. So $k'$ is algebraic over $k$. So $K$ is algebraic over $k$.///

Weak Nullstellenstaz. Let $k$ be an algebraically closed field. Let $I$ be an ideal in $k[x_1,x_2,\cdots,x_n]$. If $I$ is a proper ideal, then polynomials in $I$ have a common zero.

Proof. Let $\mathfrak{m}$ be an maximal ideal that contains $I$. Let $\phi:k[x_1,x_2,\cdots,x_n]\to K=k[x_1,x_2,\cdots,x_n]/\mathfrak{m}$ be the quotient map. $K$ is a field extension of $k$ that is generated as a $k$-algebra by $\alpha_i=\phi(x_i)$. Therefore, by the previous proposition, $K$ is algebraic over $k$. But $k$ is algebraically closed, so $K=k$. So $P=(\alpha_1,\alpha_2,\cdots,\alpha_n)$ is a point in $k^n$, and all polynomials vanish at $P$.///

Strong Nullstellensatz. Let $k$ be an algebraically closed field. Let $g,g_1,g_2,\cdots,g_m$ be elements in $k[x_1,x_2,\cdots,x_n]$. If $g$ vanishes at all common zeros of $g_1,g_2,\cdots,g_m$, then $g^p$ is contained in the ideal generated by $g_1,g_2,\cdots,g_m$ for some positive integer $p$.

Proof. The case when $g=0$ is trivial. Now we assume $g\neq0$. By adding a new variable $x_0$, we can construct a set of polynomials in $k[x_0,x_1,x_2,\cdots,x_n]$: $1-x_0g,g_1,g_2,\cdots,g_m$. We observe that this new set of polynomials do not have a common zero. By Weak Nullstellensatz, there is a set of polynomials $r,r_1,r_2,\cdots,r_m$ in $k[x_0,x_1,x_2,\cdots,x_m]$ such that
$$r(1-x_0g)+r_1g_1+r_2g_2+\cdots+r_mg_m=1$$
Consider this equation and all elements in $k(x_1,x_2,\cdots,x_m)[x_0]$, that is, a polynomial ring of a single variable $x_0$ with coefficients in the field of fraction of $k[x_1,x_2,\cdots,x_n]$. Evaluate the equation at $$x_0=g^{-1}$$.
We have
$$r_1(g^{-1})g_1+r_2(g^{-1})g_2+\cdots+r_m(g^{-1})g_m=1$$
By multiplying $g^p$ for some sufficiently large $p$, we can clear the power of $g$ in the fraction of this equation
$$g^pr_1(g^{-1})g_1+g^pr_2(g^{-1})g_2+\cdots+g^pr_m(g^{-1})g_m=g^p$$
That is,
$$g^pr_1(g^{-1}),g^pr_2(g^{-1}),\cdots,g^pr_m(g^{-1})$$
are polynomials in $k[x_1,x_2,\cdots,x_m]$. Therefore, $g^p$ is in the ideal generated by $g_1,g_2,\cdots,g_m$.///
