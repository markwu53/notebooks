A ring has two binary operations, addition and multiplication. With these two operations, we can write arithmetic expressions in terms of ring elements. By asking the inverse question - find elements that make an arithmetic expression equal to zero - we have the arithmetic equation or *algebraic equation*.

The solution to an algebraic equation may not exist in the original ring. But it may exist in some extension of the ring. The extension can be an imaginary construction. After all, the question can be expressed in the minimum space that we can ask such a question, which is a polynomial ring with as many variables as in the question, together with the question, which is a polynomial, itself!

A special class of question is to ask if an element has an inverse. This is like ask to solve an equation of $xy=1$, or consider the polynomial $xy-1$ in the polynomial ring with two variables $x$ and $y$.

This question has already been solved, which leads to the concept of field of fraction of a ring. This is a base of solving other algebraic equations.

Let $R$ be a commutative ring.

Let $R$ be an integral domain.

Let $\mathcal{I}$ be the set of ideals of $R$. Each $I\in\mathcal{I}$ can be viewed as a $R$-module. $R$ itself can be viewed as an $R$-module, and $\mathcal{I}$ is the space of $R$-submodules of $R$.

Each ideal is also an $R$-subalgebra. We don't call it a subring, because it doesn't contain the identity.

Let $R^0=R\backslash\{0\}$. It's closed under multiplication, and has an identity.

A *multiplicative set* is a subset of $R^0$ that is closed under multiplication and has an identity.

Remark. $R^0$ is a $\mathbb{Z}$-module. Usually it's not a finitely generated $\mathbb{Z}$-module. Every multiplicative set is a submodule of $R^0$.

Remark. Given two multiplicative sets, we can form new multiplicative set by intersection, and by direct sum of $\mathbb{Z}$-submodule. But tensor product is not a way to form new multiplicative set.

Let $\mathcal{M}$ be the set of multiplicative set of $R$.

Remark. A principal $\mathbb{Z}$-submodule of $R^0$ is a multiplicative set generated by a single element.

Remark. The complement of a prime ideal is a multiplicative set.

Definition. $R/R^0$ is called the *field of fraction* of $R$.

$F$ is also an $R$-module. Let $\mathcal{J}$ be the space of $R$-submodules of $F$. Each element of $\mathcal{J}$ is called a *fractional ideal*. $\mathcal{I}\subseteq\mathcal{J}$. Among fractional ideals, a usual ideal, an element in $\mathcal{I}$, is also called an *integral ideal*.

Remark. A fractional ideal either contains $1$ or not contains $1$. If it does not contain $1$, the it's an integral ideal, or a usual ideal. If it contains $1$ then it contains the whole $R$ as a subset. If it's not $R$ itself, then it's a *pure* fractional ideal.

A fractional ideal may not be a $R$-subalgebra. A $R$-subalgebra generated by a pure fractional ideal is a subring of $F$, the field of fraction of $R$.

We can define several binary operations on $\mathcal{I}$. Let $\mathfrak{a}$ and $\mathfrak{b}$ be two ideals.

$\mathfrak{a}\cap\mathfrak{b}$ is also an ideal.

Define $\mathfrak{a}+\mathfrak{b}$ to be the ideal generated by $\mathfrak{a}$ and $\mathfrak{b}$, which is $\mathfrak{a}\vee\mathfrak{b}$, the *join* of $\mathfrak{a}$ and $\mathfrak{b}$ in the lattice $\mathcal{I}$, or the least upper bound of $\mathfrak{a}$ and $\mathfrak{b}$, or the intersection of all upper bounds of $\mathfrak{a}$ and $\mathfrak{b}$.

Define $\mathfrak{a}\oplus\mathfrak{b}$ to be the direct sum of $\mathfrak{a}$ and $\mathfrak{b}$ as $R$-submodules, which is $\{a+b: (a,b)\in \mathfrak{a}\times\mathfrak{b}\}$.

Remark. $\mathfrak{a}+\mathfrak{b}=\mathfrak{a}\oplus\mathfrak{b}$.

Define $\mathfrak{a}\otimes\mathfrak{b}$, also simply written as $\mathfrak{a}\mathfrak{b}$, to be the tensor product of $R$-subalgebra, which has elements of the form $\sum a_ib_i$ of finite sum.

Remark. $\mathfrak{a}\mathfrak{b}\subseteq\mathfrak{a}\cap\mathfrak{b}$, and sometimes $\mathfrak{a}\mathfrak{b}\subsetneq\mathfrak{a}\cap\mathfrak{b}$

Remark. With $\oplus$ and $\otimes$, $\mathcal{I}$ is a semiring with $R$ itself being identity.

We can also define similar binary operations for fractional ideals. Let $\mathfrak{f}$ and $\mathfrak{g}$ be two fractional ideals.

We can define $\mathfrak{f}+\mathfrak{g}$ or $\mathfrak{f}\oplus\mathfrak{g}$, the $R$-submodule generated by $\mathfrak{f}$ and $\mathfrak{g}$. $\mathfrak{f}\cap\mathfrak{g}$ is also a $R$-submodule.

$\mathfrak{f}\otimes\mathfrak{g}$ is defined to be the tensor product of $R$-submodules as a subet of a $R$-subalgebra. To spell out, all elements of $\mathfrak{f}\otimes\mathfrak{g}$ have the form $\sum f_ig_i$ of finite sum.

Remark. With $\oplus$ and $\otimes$, $\mathcal{J}$ is a semiring with $R$ itself being identity.

Remark. Principle $R$-submodules are all units in $\mathcal{J}$, and they form a group. It's a subgroup of the group of units of $\mathcal{J}$.

Definition. The quotient group of the group of units of $\mathcal{J}$ modulo the subgroup of principal $R$-submodules is called the *class group* of $R$.

Let $\mathcal{R}$ be the set of subrings of the field of fractions $F$ of $R$ that contains $R$ as a subring. Every element in $\mathcal{R}$ is a $R$-subalgebra of $F$.

Intersection of two subrings is still a subring. The subring generated by two subrings is equal to the tensor product of two subrings viewed as two $R$-subalgebras, which is also equal to the "direct" product of two subrings, that is, every element has the form $s_1s_2$ with $s_1$ in the first subring and $s_2$ in the second subring.

Remark. $\mathcal{I},\mathcal{M},\mathcal{J},\mathcal{R}$ all have a module structure. $\mathcal{I}$ and $\mathcal{J}$ are $R$-submodules. $\mathcal{M}$ are $\mathbb{Z}$-modules. $\mathcal{R}$ are $R$-submodules and also $R$-subalgebras. They are all partially ordered sets with inclusion as the relation. They all have lattice structure, with intersection be the *meet* and element generated by two elements as the *join*. Now we study the relation among these lattice structures.

Define a topology on $\mathcal{I}$ by such that for any $I\in\mathcal{I}$ define the subset of $\mathcal{I}$ of ideals that contains $I$ to be a closed set. This is called the *Zaraski topology* on $\mathcal{I}$.

For an ideal $I\in\mathcal{I}$, let $T_c(I)$ denote the closed set in $\mathcal{I}$ related to $I$. $T_c$ is a bijection from $\mathcal{I}$ to the set of closed sets of $\mathcal{I}$.

Modeled by $\mathbb{Z}$, we see that the structure of $\mathcal{I}$ is like a cartesian product of the set of prime numbers and the set of natural numbers, or a bundle with the set of prime numbers as base and the set of natural numbers as fiber. An element or an ideal is a section of the bundle with finite support. The Zaraski topology is a product topology of the restriction on the base and the restriction on the fiber, which in this case are both finite complement topology.

Definition. Let $Spec(R)$ be the set of prime ideals of $R$. The *Zaraski topology* on $Spec(R)$ is the Zaraski topology on $\mathcal{I}$ restricted on $Spec(R)$.

For $I\in\mathcal{I}$, define $L:\mathcal{I}\to\mathcal{R}$ by $L(I)=\cap_f R/M_f$ where $M_f$ is the multiplicative set generated by an element $f\in I$. 

Remark. The function $L$ is called *localization*. We can localize a ring at an ideal to obtain a subring in the field of fraction of the ring. We can localize at an element of the ring, which is just localize at the principal ideal generated by the element.

Since $T_c$ is a bijection from $\mathcal{I}$ to the Zaraski topology on $\mathcal{I}$ (We use $T_C$ to denote the mapping in both direction). $T_C$ and $L$ define a function from the Zaraski topology of $\mathcal{I}$ to $\mathcal{R}$.

Let $\phi$ denote the restriction map from the Zaraski topology of $\mathcal{I}$ to the Zaraski topology of $Spec(R)$.

We will show that $T_C, L$ and $\phi$ induce a map from the Zaraski topology of $Spec(R)$ to $\mathcal{R}$, also called *localization* and denoted by $L$, such that $L(\phi(T_C(I)))=L(I)$ for any $I\in\mathcal{I}$. This means if $I_1$ and $I_2$ are such that $\phi(T_C(I_1))=\phi(T_C(I_2))$, then $L(I_1)=L(I_2)$, or if the closed set defined by $I_1$ and $I_2$ contain the same set of prime ideals, then they obtain the same subring.

Remark. For a proper ideal $I$, $\phi(T_C(I))$ is nonempty, since every proper ideal is contained in a prime ideal.

For an proper ideal $I$ define $PR(I)$ to be the intersection of the prime ideals in $\phi(T_C(I))$.

A *radical ideal* $I$ is an ideal such that if $f^n\in I$ then $f\in I$. For an ideal $I$, the *radical closure* of $I$, denoted $R(I)$, is the radical ideal generated by $I$, which is the smallest radical ideal contains $I$, or the intersection of all radical ideals containing $I$.

We will show that $L(I)=L(R(I))$ and $R(I)=PR(I)$.

First $I\in R(I)$ so $L(R(I))\subseteq L(I)$. For an element $f\in R(I)$ and $f\notin I$, $f^n\in I$ for some $n$. And $L(P_f)=L(P_{f^n})$, where $P_f$ is the principal ideal generated by $f$. So $L(I)\subseteq L(R(I))$.

Now we show $R(I)=PR(I)$. This is surprisingly nontrivial. First it's easy to see a prime ideal is radical, and intersection of radical ideals is radical. So $R(I)\subseteq PR(I)$. Next we consider the quotient map $R\to R/I$. $I$ becomes the zero ideal. $R(I)$ becomes the *nilradical*, which is the radical of the zero ideal, and prime ideals that contain $I$ are still prime ideals, and they are the only prime ideals in the image. We will show in the image, the intersection of prime ideals is the nilradical, and thus in the inverse image $PR(I)\subseteq R(I)$. Suppose not. Then we have an element $f\in Q=R/I$ that is contained in every prime ideal but $f^n\neq 0$ for all $n$. Then $f$ generate a multiplicative set $S$, and the ring $Q/S$ has a maximal ideal and thus a prime ideal that does not contain $f$ because $f$ is a unit in this ring. Look at the inclusion map $Q\to Q/S$, the prime ideal in $Q/S$ pull back to a prime ideal in $Q$ that does not contain $f$. Contradiction.

Remark. Viewing $\mathcal{I}$ as a bundle over the base $Spec(R)$, radical captures the structure in the fiber. Modulo the radicals, the structure of $\mathcal{I}$ is determined by $Spec(R)$.

Theorem. The localization function is well defined on the Zaraski topology (ope
n sets) of $Spec(R)$, and it's order reversing with the inclusion as the order relation.

Remark. Viewing the poset $Spec(R)$ as a category, and $\mathcal{R}$ as a categ
ory, the localization function is contravariant functorial and is called a *sheaf*.

Given a filtration of open sets around a point in $Spec(R)$, the direct limit in $\mathcal{R}$ is the *stalk* of the sheaf at a prime ideal.
