### Scipy -- Special Functions

Bessel functions are a family of solutions to Besselâ€™s differential equation:

$$
x^2\frac{d^2y}{dx^2} + x\frac{dy}{dx} + (x^2 - \alpha^2)y = 0
$$

We need to import Numpy and Scipy:

```python
import numpy as np
from scipy import special
```

Then we define a function:

```python
def drumhead_height(n, k, distance, angle, t):
   kth_zero = special.jn_zeros(n, k)[-1]
   return np.cos(t) * np.cos(n*angle) * special.jn(n, distance*kth_zero)
```

Then for a grid of points $(x,y)$, we calculate $z=f(x,y)$, where $f$ is the Bessel function of order $0$:

```python
theta = np.r_[0:2*np.pi:50j]
radius = np.r_[0:1:50j]
x = np.array([r * np.cos(theta) for r in radius])
y = np.array([r * np.sin(theta) for r in radius])
z = np.array([drumhead_height(1, 1, r, theta, 0.5) for r in radius])
```

To plot the graph, we need to import Matplotlib:

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

Don't forget to add the magic line:

```python
%matplotlib inline
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.jet)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
