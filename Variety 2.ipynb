Let $R$ be a commutative ring.

Definition. Let $I\in R$ be an ideal.
* $I$ is $\textbf{radical}$ if $f^n\in I\implies f\in I$.
* The $\textbf{radical closure}$ of $I$, denoted $R(I)$, is the radical ideal generated by $I$, or the smallest radical ideal contains $I$, or the intersection of radical ideals that contain $I$.
* $R(0)$ is called the $\textbf{nilradical}$.

Remark.
* A radical ideal is like removing repeated factors. In a unique factorization domain, a principle ideal is radical if and only if the generator element does not have repeated factors.
* $R(R(I))=R(I)$.
* A prime ideal is radical.
* The intersection of radical ideals is radical.
* Nilradical is equal to the intersection of all prime ideals.
* A radical ideal is equal to the intersection of some collection of prime ideals.

Let $R$ be an integral domain.

Definition. Let $R^0=R\backslash\{0\}$.
* A $\textbf{multiplicative set}$ is a submonoid of $R^0$.
* A multiplicative set generated by a single element $f\in R^0$ is denoted $M(f)$.

Remark.
* A multiplicative set $M$ is a module over the semiring $\mathbb{Z}_+$, the semiring of nonnegative integers, where an element $n\in\mathbb{Z}_+$ acts on an element $f\in M$ by $n.f=f^n$.
* The compliment of a prime ideal is a multiplicative set.

Definition.
* The $\textbf{field of fraction}$ of $R$ is $R/R^0$.
* A $\textbf{ring of fraction}$ of $R$ is $R/S$ where $S$ is a multiplicative set.
* Let $I$ be an ideal. The $\textbf{ring of fraction}$ of $I$ is $F(I)=\cap_{f\in I} R/M(f)$.
* The $\textbf{localization}$ of $R$ at a prime ideal $\mathfrak{p}$ is $R/S$ where $S$ is the multiplicative set $R\backslash\mathfrak{p}$.

Remark.
* The field of fraction of $R$ is a $R$-module as well as a $R$-algebra.
* A ring of fraction of $R$ is a subring, submodule, subalgebra of the field of fraction of $R$.
* $F(I)=F(R(I))$, so $F$ is injective on the set of radical ideals.
* $I\subseteq J\implies F(J)\subseteq F(I)$.
* $F(I)=[F(J)]_{I\subseteq J}$, where square bracket is the generator notation.
* $F(I)=[F(\mathfrak{p})]_{I\subseteq\mathfrak{p}}$
* $F(I)=\cap_{J\subseteq I} F(J)$.

Definition.
* Let $\mathcal{I},R\mathcal{I},Spec(R)$ be the set of ideals, radical ideals, prime ideals of $R$. Let $P\mathcal{I}$, $PR\mathcal{I}$, $PSpec(R)$ be the power set of corresponding sets.
* $Spec(R)\to R\mathcal{I}\to\mathcal{I}$ has inclusion maps denoted by $\alpha$ and $\beta$. Let $\alpha^{-1}$ and $\beta^{-1}$ be the inverse image maps between corresponding power sets.
* $\mathcal{I}\to R(\mathcal{I})$ has a map $R$ of taking radical closure. Let $R^{-1}$ be the inverse image map between corresponding power sets.
* Define a function $T_c:\mathcal{I}\to P(\mathcal{I})$ by $T_c(I)=\{J\in \mathcal{I}:I\subseteq J\}$.

Remark.
* $T_c$ is injective.
* $T_c\circ\beta\circ R=T_c$.
* $\cap (\cap_i T_c(I_i))=[I_i]_i=\coprod I_i$, where $[I_i]_i$ is the ideal generated by $I_i$, and $\coprod I_i$ is the coproduct or direct sum of $I_i$.
* $I\cap J=\cap(T_c(I)\cup T_c(J))$.
* $\cap_i T_c(I_i)=T_c(J)$, where $J=\cap (\cap_i T_c(I_i))=[I_i]_i=\coprod I_i$.
* $T_c(I)\cup T_c(J)\subseteq T_c(K)$, where $K=I\cap J=\cap(T_c(I)\cup T_c(J))$.
* $\alpha$ is prime ideal $\implies$ $(I\cap J\subseteq\mathfrak{a}$ $\implies$ $I\subseteq\mathfrak{a}$ or $J\subseteq\mathfrak{a})$.
* $\bar{T}_c(I)\cup \bar{T}_c(J)=\bar{T}_c(K)$ where $\bar{T}_c=T_c\circ\beta^{-1}\circ\alpha^{-1}$.
* $\beta\circ T_c\circ\beta^{-1}\circ\alpha^{-1}$ is injective.

Definition. Define a topology on $Spec(R)$ by asserting $\{\mathfrak{p}\in Spec(R):I\subseteq\mathfrak{p}\}$ to be closed for any $I\in\mathcal{I}$. This is called the $\textbf{Zaraski topology}$ on $Spec(R)$.

Remark.
* The method almost define a topology on $\mathcal{I}$, except the closed sets are not closed under finite union.
* The difference between $\mathcal{I}$ and $Spec(R)$ is the radicals. In a sense, $\mathcal{I}$ can be viewed as a bundle over $Spec(R)$. The fiber is the repeated factors. It has a relatively simple structure. The base space $Spec(R)$ carrys most information of $\mathcal{I}$.
* Radical ideals are in one to one correspondence with closed sets of $Spec(R)$, therefore also in one to one correspondence with open sets of $Spec(R)$.

Definition. Let $(X,\mathcal{T}_X)$ be a topological space. A $\textbf{presheaf}$ of rings over $X$ is an order-reversing function $\mathcal{O}_X:\mathcal{T}_X\to Ring$, where the order on $\mathcal{T}_X$ and $Ring$ is given by inclusion.

Remark.
* A partially order set can be viewed as a cetegory. So $\mathcal{O}_X$ is a contravariant functor from $\mathcal{T}_X$ to $Ring$.
* The function from the Zaraski topology on $Spec(R)$ to $Ring$, established via radical ideals, is a presheaf.
* Let $X=Spec(R)$. Let $\mathcal{O}_X$ be the presheaf as above. Let $\cdots \subseteq U_{i+1}\subseteq U_{i}\subseteq\cdots$ be a filtration of open sets around a prime ideal $\mathfrak{p}\in Spec(R)$. The direct limit of $\mathcal{O}_X(U_i)$ is equal to the localization of $R$ at $\mathfrak{p}$.

Let $k$ be an algebraically closed field. Let $R=k[x_1,x_2,\cdots,x_n]$.

Remark.
* For any field $k$, $k[x]$ is a principal ideal domain, and thus a unique factorization domain.
* If $R$ is a unique factorization domain, then $R[x]$ is also a unique factorization domain.
* $k[x_1,x_2,\cdots,x_n]$ is a unqiue factorization domain.
* In a unique factorization domain, irreducible elements generate prime ideals.
* $k[x_1,x_2,\cdots,x_n]$ is noethrian, i.e., every ideal is finitely generated.
* A quotient ring of $k[x_1,x_2,\cdots,x_n]$ is a finitely generated $k$-algebra. If the kernel is a radical ideal, then the quotient ring is a reduced $k$-algebra, i.e., zero nilradical.

Definition. Let $I$ be an ideal of $k[x_1,x_2,\cdots,x_n]$.
* An $\textbf{algebraic set}$, denoted $V(I)$, is the set of points $x\in k^n$ such that $f(x)=0$ for all $f\in I$.
* An $\textbf{algebraic variety}$ is an *irreducible* algebraic set, which is an algebraic set that is not a union of two proper algebraic subsets.
* $R/R(I)$ is called the $\textbf{coordinate ring}$ of an algebraic set $V=V(I)$, denoted $k[V]$.

Remark.
* $V(I)=V(R(I))$.
* $k^n$ is an algebraic set corresponding to the zero ideal.
* An algebraic variety corresponds to a prime ideal.
* An algebraic variety can contain sub-varieties, but cannot be *finite* union of sub-varieties.
* Algebraic sets and radical ideals are in one to one correspondence, by Hilbert Nullstallenstaz.
* The Zaraski topology on $Spec(R)$ extends to the Zaraski topology on the set of algebraic sets. The connection is via radical ideals.
* The Zaraski topology on the set of algebraic sets also gives the Zaraski topology on $k^n$, since $k^n$ is the containing algebraic set of every other algebraic set.
* The Zaraski topology on an algebraic set or an algebraic variety is the subspace topology of the Zaraski topology on $k^n$.

Remark.
* A polynomial $f\in k[x_1,x_2,\cdots,x_n]$ defines a function from $k^n$ to $k$ by evaluating.
* Two polynomials $f$ and $g$ define the same function on an algebraic set $V=V(I)=V(R(I))$ if and only if $f=g$ modulo $R(I)$.
* An element in the coordinate ring $k[V]$ defines a function from an algebraic set $V$ to $k$.
* Two polynomials $f$ and $g$ define a function $f/g$ on the open subset of $k^n$ where $g$ is not zero.

Definition. Let $V$ be an algebraic set. Let $k[V]$ be its coordinate ring.
* A $\textbf{rational function}$ on $V$ is of the form $f/g$ for $f,g\in k[V]$.
* A $\textbf{regular function}$ on an open subset $U$ of $V$ is a function locally of the form $f/g$ everywhere on $U$ for some $f$ and $g$ in $k[V]$.

Remark.
* A rational function $f/g$ is not really a function. It's only a function wherever $g$ is not zero. So it defines a function on a open set.
* A regular function $f$ on an open set $U$, when restricted on an open subset $W\subset U$, becomes a regular function on $W$.
* The set of regular functions on an open set $U$ is a ring.
* Let $\mathcal{O}_V(U)$ be the ring of regular functions on $U$. Then $\mathcal{O}_V$ is a presheaf of ring of regular functions over $V$ (with Zaraski topology).
* Let $X=Spec(k[V])$. We can identify the Zaraski topology on $X$ and the Zaraski topology on $V$, i.e., $\mathcal{T}_X=\mathcal{T}_V$. There is an ring isomorphism between $\mathcal{O}_X(U)$ and $\mathcal{O}_V(U)$ for every open subset $U$ of $V$. Or, there is a natural isomorphism between two functors $\mathcal{O}_X$ and $\mathcal{O}_V$.

Definition.
A $\textbf{ringed space}$ is a triple $(X,\mathcal{T}_X,\mathcal{O}_X)$ where $(X,\mathcal{T}_X)$ is a topological space and $\mathcal{O}_X$ is a presheaf of rings over $X$. 

Remark. Let $V$ be an algebraic set. Let $\mathcal{T}_V$ be the Zaraski topology on $V$. Let $\mathcal{O}_V$ be the presheaf of ring of regular functions on $V$. Then $(V,\mathcal{T}_V,\mathcal{O}_V)$ is a ringed space.

Definition. Let $V$ and $W$ be two algebraic sets, with presheaf of ring of regular functions $\mathcal{O}_V$ and $\mathcal{O}_W$. A function $\phi:V\to W$ is $\textbf{regular}$ if $\phi$ is continuous and $f\circ\phi \in \mathcal{O}_V(\phi^{-1}(U))$ for any $f\in\mathcal{O}_W(U)$.

Remark. A regular function between two algebraic sets is a natural transformation between two functors $\mathcal{O}_W$ and $\mathcal{O}_V\circ\phi^{-1}$ where the natural transformation $\alpha$ is given by $\alpha_U(f)=f\circ\phi$.
