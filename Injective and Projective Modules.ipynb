Let $R$ be a commutative ring.

Definition.
* An $R$-module $I$ is $\textbf{injective}$ if as a submodule of any module it is a direct summand of that module.
* An $R$-module $I$ is $\textbf{injective}$ if as a submodule of any module $M$ there exists a submodule $N$ of $M$, such that $N\cap I=\{0\}$ and $N\oplus I=M$.
* An $R$-module $I$ is $\textbf{injective}$ if for any pair of module and submodule, any homomorphism on the submodule with values in $I$ always extends to the module.
* An object $I$ in a category $\mathcal{C}$ is $\textbf{injective}$ if for any object pair with a *monomorphism* between them $\iota:M\hookrightarrow N$, any morphism $\phi:M\to I$ induces a morphism $\psi:N\to I$ such that $\phi=\psi\circ\iota$.

Remark.
* Any module can be a direct summand as a submodule of another module. But the definition asserts an injective module can *only* be a direct summand. In what situation a module is a submodule of another module but not a direct summand? For example, $\mathbb{Z}_2$ in $\mathbb{Z}_4$ as $\mathbb{Z}$-modules. $3\mathbb{Z}$ in $\mathbb{Z}$ as $\mathbb{Z}$-modules. It seems that the submodule in this case is too *coarse*. There are *finer* elements that can *fill* in it to make a larger submodule. This notion can be pursued in two senses.
* If all *finer* elements are already contained in a module, makes it a *fully divisable* module, then there is no way to fill it more, any element can only adjoin to it in *other direction*, make it a direct summand inside any module.
* If an element can in some setting prevent a module being a direct summand, then adjoining this element to the module to make a larger module can remove the obstruction of the element preventing the module being a direct summand, so the new module is closer to being a direct summand in any setting.
* The difficulty in thinking injective module is that the definition is not in a *particular* setting, it is stated in *any* setting.
* The difficulty in understanding the third and fourth definition is that, it mixes the easy case with the real case. The easy case is when the submodule is already a direct summand, which we sometimes implicitly think that way. In the easy case, we can simply set the values on the other part of the direct summand to be zero, or freely assign any value on the *generators* of the other part of the direct summand. Removing this easy case, the problem falls back to the real case, which is the same as in other definitions.

Remark.
* A ring provides numbers and units of measurements. A module provides an arrangement of multiple sets of measurable objects. A module being injective depends on the arrangement as well as the ring.
* An injective module is usually large, because it contains all finer objects to become fully divisable, or it contains elements that may have intersections with it to become maximal in essential extension.

Remark.
* The direct sum of injective modules is injective.
* If $E$ is an injective module and $S$ an $R$-algebra, then $Hom_R(S,E)$ is an injective $S$-module.

Definition. If an injective module has no proper injective submodule then it is an $\textbf{indecomposable}$ injective module.

Remark.
* At a first glance, injective modules seem to be ubiquitous, but it's not. We get the experience from vector spaces over a field -- any subspace of a vector space is a direct summand -- but this is far from true for a general ring.
* For example, consider the injective $\mathbb{Z}$-module homomorphism $\iota:\mathbb{Z}_2\to\mathbb{Z}_4$ such that $1\mapsto 2$, the submodule $\{0,2\}$ is not a direct summand of $\mathbb{Z}_4$.
* Another example, consider the injective $\mathbb{Z}$-module homomorphism $\iota:\mathbb{Z}\to\mathbb{Z}$ such that $1\mapsto 3$, the submodule $3\mathbb{Z}$ is not a direct summand of $\mathbb{Z}$.
* The reason for the failure is that, to be a direct summand, a submodule $I$ cannot intersect nontrivially any submodule generated by elements not in $I$. If $I$ is not a direct summand, then there exists an element $x$ not in $I$ such that the principal submodule generated by $x$ intersects $I$ nontrivially. In this case, the submodule $J$ generated by $I$ and $x$ is closer to being a direct summand.

Definition.
* $E$ is an $\textbf{essential submodule}$ of $M$ if for any $x\in M$ not in $E$, $E\cap (x)
\ne \{0\}$ where $(x)$ is the principle submodule generated by $x$.
* An injective module homomorphism $\iota:M\to N$ is an $\textbf{essential extension}$ if $\iota(M)$ is an essential submodule of $N$. In this case, we also say $N$ is an essential extension of $M$.
* If $E$ has no proper essential extension, then $E$ is $\textbf{essential extension closed}$.
* Let $M$ be a module. The smallest module that is essential extension closed and contains $M$ is called an $\textbf{essential extension closure}$ of $M$.

Remark.
* Every module is an essential submodule of its own.
* An essential extension closed module is an essential extension closure of its own.
* Every module has an essential extension closure.
* The essential extension closure is unique.

Remark.
* An $R$-module is injective if and only if it's essential extension closed.

Definition. Let $I(M)$ be the samllest injective module that contains $M$. It is also the essential extension closure of $M$.

Defintion.
* Given an $R$-module $M$, let $M_0=M$, $I_0=I(M_0)$, $M_1=I_0/M_0, I_1=I(M_1), \cdots$, we obtain an long exact sequence:
$$0\to M_0\to I_0\to M_1\to I_1\to M_2\to I_2\to\cdots$$
Combining adjacent injection and projection to remove $M_{*}$, we get another long exact sequence:
$$0\to M\to I_0\to I_1\to I_2\to\cdots$$
This is called the $\textbf{minimal injective resolution}$ of $M$.

Remark.
* $M\to I_0$ is an injection. $I_n\to I_{n+1}$ are not injections.

Definition.
* A module $P$ is $\textbf{projective}$ if it's a direct summand of a free module.
* A module $P$ is $\textbf{projective}$ if for any surjective homomorphism $\phi:M\to P$ we have $M\cong Ker(\phi)\oplus P$.
* A module $P$ is $\textbf{projective}$ if for any pair of module with a surjective homomorphism $f:M\to N$, a homomorphism $\phi:P\to N$ always lifts to a homomorphism $\psi:P\to M$, that is, $\phi=f\circ\psi$.

Remark.
* The various forms of definition are equivalent. (1)=>(3)=>(2)=>(1).
* (1)=>(3). Consider the generators of the free module. On the one hand they project to $P$ and map to $N$ by $\phi$. On the other hand, we can freely assign their values in $M$, and them maps to $N$ by the surjection $f$, such that they agree with the other route.
* (3)=>(2). By letting $N=p$ and $\phi$ be the identity map, $P$ has a homomorphism maps to a section in $M$. $Ker(f)$ cannot intersect this section which is a submodule. So $M$ is the direct sum of the $Ker(f)$ and the section submodule.
* (2)=>(1). By leeting $M$ to be the free module. The projection to $P$ splits. So it's a direct summand.

Remark.
* The free module has the *lifting* property. (3) captures this property. So a free module is always a projective module.
* On the other hand, a torsion module cannot be a direct summand of a free module, so cannot be a projective module.
* So a projective module is between free and torsion-free. A property that lives between a sufficient condition and a necessary condition, but the exact condition is hard to characterize. That signifies something deep.

Definition.
* A $\textbf{projective resolution}$ of a module $M$ is a long exact sequence of projective modules $P_n$:
$$\cdots\to P_2\to P_1\to P_0\to M\to 0$$
* A $\textbf{free resolution}$ and a $\textbf{flat resolution}$ is defined similarly.

Definition.
A $\textbf{complex}$ is a sequence of modules $C_n$ and homomorphisms $d_n:C_n\to C_{n-1}$ such that $d_{n-1}\circ d_n=0$. A complex can $\textbf{bound above}$ or $\textbf{bound below}$. If we reverse the arrows, we get $\textbf{cocomplex}$, in which case we also reverse the indexes. For a complex we can take the $\textbf{homology}$ $H_n=Ker(d_n)/Im(d_{n+1})$, and for a cocomplex we take the $\textbf{cohomology}$.

Remark.
* A long exact sequence is a complex.
* A short exact sequence is also a complex, which bound above and below.
* We can define homomorphisms between two complexes. Kernel, image, and cokerkel of a complex homomorphism can all be defined. We can even define complex of complexes. For two complex homomorphisms between same two complexes, there is a concept of a homotopy. A complex homomorphism induces a homology homorphism. Homotopic complex homomorphisms induce the same homology homomorphism.

Remark.
* Complexes form a category. The morphisms are complex homormorphims, or complex homorphisms up to homotopy. We can define functors on it.
* $-\otimes N$, $Hom(-,N)$ and $Hom(N,-)$ are all endofunctors on the category of complexes.
* Two projective resolutions of a module induce two complexes by above functors, but they induce the same homology.

Defintion.
* A covariant functor on the category of modules is $\textbf{left exact}$ if it preserves the injection on a short exact sequence, $\textbf{right exact}$ if it preserves the surjection on a short exact sequence, and is $\textbf{exact}$ if it preserves both. Similar for a contravariant functor.
* A module $M$ is $\textbf{flat}$ if the functor $M\otimes -$ is exact.

Remark.
* A functor on the category of modules is usually also a functor on the category of complexes.
* $M\otimes -$ is left exact for a module $M$, and exact for a projective module $M$.
* $Hom(-,M)$ is left eact for a module $M$, and exact for an injective module $M$.
* A free module is always projective. A projective module is always flat. A flat module has no torsion.

Definition.
* $Tor_n(M,N)=H_n(P_{*}\otimes N)$ where $P_{*}$ is a projective resolution of $M$.
* $Ext^n(M,N)=H^n(Hom(M,I_{*}))$ where $I_{*}$ is an injective resolution of $N$.
* $Ext^n(M,N)=H^n(Hom(P_{*},N))$ where $P_{*}$ is a projective resolution of $M$.
