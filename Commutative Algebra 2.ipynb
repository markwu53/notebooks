Let $R$ be a commutative ring. (always with identity)

Let $R$ be an integral domain.

Notations:
* Let $\mathcal{I}$ be the category of ideals of $R$.
* Let $\mathcal{R}$ be the category of radical ideals of $R$.
* Let $\mathcal{F}$ be the category of subrings of the field of fractions of $R$ which are also $R$-subalgebras.
* Let $\mathcal{Q}$ be the category of quotient rings of $R$.
* Let $\mathcal{M}$ be the category of multiplicative sets of $R$. A multiplicative set is a $\mathbb{Z}_+$-module where $\mathbb{Z}_+$ is the semiring of the nonnegative integers. Module addition is defined by multipication in $R$. $\mathbb{Z}_+$ acts on $R$ by $n\cdot r=r^n$.
* Let $R^0=R\backslash\{0\}$. It is the biggest multiplicative set of $R$.
* Let $M:R^0\to\mathcal{M}$ be a function such that $M(f)$ is the multiplicative set generated by $f\in R^0$.
* Let $\mathcal{J}$ be the category of fractional ideals of $R$, which are $R$-submodules of the field of fractions of $R$.
* Let $Spec(R)$ be the set of prime ideals of $R$.

Facts:
* $\mathcal{I}\to\mathcal{R}$ has a functor by taking radical closure of an ideal.
* $\mathcal{I}\to\mathcal{F}$ has a functor $I\mapsto\cap_f R/M(f)$.
* $\mathcal{I}\to\mathcal{Q}$ has a functor by taking quotient.

Let $k$ be an algebraically closed field.

Let $R=k[\textbf{x}]=k[x_1,x_2,\cdots,x_n]$ be the polynomial ring of $n$ variable over $k$.
