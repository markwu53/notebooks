Gauss Lemma. Let $R$ be an unique factorization domain. If $r_1,r_2$ are primitive in $\mathbb{R}[x]$, then $r_1*r_2$ is also primitive. Proof. We have shown this in another article for $\mathbb{Z}$. The argument is the same for a general u.f.d. because in a u.f.d. an irreducible element is exactly like a prime number in $\mathbb{Z}$.

Corrolary. Let $R$ be an u.f.d. and $F$ be the field of fraction of $R$. If an element $r$ is irreducible in $R[x]$ then it's also irreducible in $F[x]$.

Theorem. Let $R$ be an unique factorization domain. Then $R[x]$ is also an unqiue factorization domain.

Proof. Let $F$ be the field of fraction of $R$. Let $a$ be an element in $R[x]$. If $a$ is not primitive, then there is an unqiue element $r$ in $R$ (up to modulo units) such that $a=rp$ such that $p$ is a primitive polynomial in $R[x]$. Any factorization of $a$ in $R[x]$ will have a factorization of $r$ in $R$. But factorization in $R$ is unique. So to prove the original statement, it suffices to prove it for an primitive element $a$ in $R[x]$.

Let's say $a$ has two factorization in $R[x]$ (into irreducibles in $R[x]$). By Gauss Lemma, any irreducible in $R[x]$ is also irreducible in $F[x]$, so the factorization of $a$ in $R[x]$ is also a factorization in $F[x]$. But $F[x]$ is a principle ideal domain (p.i.d.) and therefore a u.f.d., so the factorization of $a$ in $F[x]$ is unique. Therefore, the two factorization of $a$ in $F[x]$ have to be the same up to fractions in $F$. Consider two factors. They are the same up to a fraction. But they are already primitive (with coefficients in $R$). So the fraction is actually a unit in $R$. So they are the same elements in $R[x]$ (up to a unit in $R$). ///

In a principal ideal domain, an ideal generated by an irreducible element is a prime ideal. Proof. Let $a$ be an irreducible element. Consider the maximal ideal that contains $a$. But assumption it must be a principal ideal, so let it be $(b)$, then $a=rb$, but $a$ is irreducible, so $r$ is a unit, therefore, $(b)=(a)$ and is maximal, thus it's prime.
