Definition. Let $A$ be a commutative ring.
* An $A$-$\textbf{ring}$ is a ring extension of $A$.
* An $A$-$\textbf{ring homomorphism}$ is a ring homomorphism that fixes $A$.
* An $A$-$\textbf{ring ideal}$ is the kernel of an $A$-ring homomorphism.

Remark.
* An $A$-ring is a ring $B$ that contains $A$ as a subring, or a ring extension of $A$.
* Let $B_1$ and $B_2$ be two $A$-rings. Let $\iota_1:A\to B_1$ and $\iota_2:A\to B_2$ be the inclusion maps. An $A$-ring homomorphism $\phi:B_1\to B_2$ is a ring homomorphism such that $\phi\circ\iota_1=\iota_2$.
* Let $B$ be an $A$-ring. An $A$-ring ideal of $B$ is an ideal of $B$ generated by a set of elements in $B\backslash A$.
* An $A$-ring is an $A$-algebra that has an identity.
* A proper ideal of $A$ is an $A$-algebra but not an $A$-ring.
* A ring of fraction, that is, $S^{-1}A$ where $S$ is a multiplicative set, is an $A$-ring.

Definition. Let $A$ be an integral domain.
* A $\textbf{monic polynomial}$ in $A[x]$ is a polynomial with leading coefficient $1$.
* Let $B$ be an $A$-ring. Given $\alpha\in B$, define a $A$-ring homomorphism $\phi_\alpha:A[x]\to B$ by $x\mapsto\alpha$. If $\phi_\alpha$ is injective, we say $\alpha$ is $\textbf{transcendental}$ over $A$. If $\alpha$ is not transcendental, if there is a monic polynomial in the kernel of $\phi_\alpha$, we say $\alpha$ is $\textbf{integral}$ over $A$, otherwise we say $\alpha$ is $\textbf{fractional}$ over $A$. If $\alpha$ is integral over $A$, the minimum degree of the monic polynomial in the kernel of $\phi_\alpha$ is called the $\textbf{degree}$ of $\alpha$. A monic polynomial in the kernel of $\phi_\alpha$ with degree equal to the degree of $\alpha$ is called a $\textbf{minimum polynomial}$ of $\alpha$.

Remark. Let $A$ be an integral domain.
* $A[x,y]=A[x][y]=A[y][x]$. Let $\iota_x:A[x]\to A[x,y]$ be an $A$-ring embedding of $x\mapsto x$ and $\iota_y:A[x]\to A[x,y]$ be $x\mapsto y$.
* Let $p,q\in A[x]$ be two monic polynomials. Let $\omega=\iota_x(p)\iota_y(q)\in A[x,y]$.
* Let $A_p=A[x]/(p)$ where $(p)$ is the principal ideal of $A[x]$ generated by $p$. Then $A_p$ is an $A$-ring, and is a free $A$-module of rank equal to the degree of $p$. $q$ is also a monic polynomial in $A_p[x]$. Let $A_{pq}=A_p[x]/(q)$ where $(q)$ is the principal ideal of $A_p[x]$ generated by $q\in A_p[x]$. $A_{pq}$ is an $A$-ring and a free $A$-module of rank $\textrm{deg}(p)\textrm{deg}(q)$. We can similarly define $A_{qp}$. Let $A_\omega=A[x,y]/(\omega)$. We have $A_{pq}\cong A_{qp}\cong A_\omega$. $A_\omega$ is a free $A$-module with basis $\{\mu_{ij}=x^iy^j:0\le i<\textrm{deg}(p),0\le j<\textrm{deg}(q)\}$.
* Let $\phi:A[x,y]\to A_\omega$ be one of the projection map defined above. Consider the $A$-subring $A[xy]\subseteq A[x,y]$. Consider the image $\phi(A[xy])$. Let $B_N=\{x^iy^i:0\le i\le N\}$. Let $A_N$ be the $A$-submodule of $A[xy]$. Consider the image $\phi(A_N)$. We want to show that if it intersects the submodule $A\mu_{ij}$ then it has to contain all of the submodule. This is also to say the image $\phi(A_N)$ is a free $A$-submodule of $A_\omega$.
* Let $B$ be an $A$-ring. Let $\alpha\in B$ be an integral element over $A$ with a minimum polynomial $p\in A[x]$. Let $I$ be the principal ideal of $A[x]$ generated by $p$. $A[x]/I$ is a free $A$-module of rank equal to the degree of $\alpha$.
